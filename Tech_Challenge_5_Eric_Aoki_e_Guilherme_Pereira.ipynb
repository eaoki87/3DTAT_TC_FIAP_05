## Tratamento de linhas em branco e separação por ano

```python
#Retirado de funções reutilizáveis do material de apoio: https://colab.research.google.com/drive/1iq0U4o_I9dfB627xtgFGIfu7GRFahD_u#scrollTo=peIQ44N-zG5f

def cleaning_dataset(df):
  _df = df.dropna(subset=df.columns.difference(['NOME']), how='all' # executa o dropna para todas as colunas sem visualizar a coluna NOME
  _df = _df[~_df.isna().all(axis=1)] # remove linhas com apenas NaN, se tiver algum dado na linha não remove
  return _df
```

```python
#Retirado de funções reutilizáveis do material de apoio: https://colab.research.google.com/drive/1iq0U4o_I9dfB627xtgFGIfu7GRFahD_u#scrollTo=peIQ44N-zG5f

def filter_columns(df, filters: list): # adiciono no array o padrão que existe nas colunas e que não quero que tenha na saída final
    selected_columns = [True] * len(df.columns)  # Inicializa todas as colunas como True
    for index, column in enumerate(df.columns):
        if any(filter in column for filter in filters): selected_columns[index] = False
    return df[df.columns[selected_columns]]
```

```python
df_2020 = df_2020 = filter_columns(df, ['2021', '2022'])
df_2020.head()
```

```python
df_2021 = filter_columns(df, ['2020', '2022'])
df_2021.head()
```

```python
df_2022 = filter_columns(df, ['2020', '2021'])
df_2022.head()
```

## Criação de métricas de variação por indicador e por ano

```python
import pandas as pd

indicadores_media_desvio = df_consolidado.groupby('ANO')[['IAA', 'IEG', 'IPS', 'IDA', 'IPP', 'IPV', 'IAN', 'INDE']].agg(['mean', 'std'])


indicadores_media_desvio.columns = ['_'.join(col) for col in indicadores_media_desvio.columns]

# Renomear as colunas
new_column_names = {
    'IAA_mean': 'IAA_Mean',
    'IEG_mean': 'IEG_Mean',
    'IPS_mean': 'IPS_Mean',
    'IDA_mean': 'IDA_Mean',
    'IPP_mean': 'IPP_Mean',
    'IPV_mean': 'IPV_Mean',
    'IAN_mean': 'IAN_Mean',
    'INDE_mean': 'INDE_Mean',
    'IAA_std': 'IAA_Std',
    'IEG_std': 'IEG_Std',
    'IPS_std': 'IPS_Std',
    'IDA_std': 'IDA_Std',
    'IPP_std': 'IPP_Std',
    'IPV_std': 'IPV_Std',
    'IAN_std': 'IAN_Std',
    'INDE_std': 'INDE_Std',
}
indicadores_media_desvio = indicadores_media_desvio.rename(columns=new_column_names)

#Criar coluans de média com adição e subtração de 1 desvio padrão
for column in ['IAA', 'IEG', 'IPS', 'IDA', 'IPP', 'IPV', 'IAN', 'INDE']:
    indicadores_media_desvio[f'{column}_Plus_Std'] = indicadores_media_desvio[f'{column}_Mean'] + indicadores_media_desvio[f'{column}_Std']
    indicadores_media_desvio[f'{column}_Minus_Std'] = indicadores_media_desvio[f'{column}_Mean'] - indicadores_media_desvio[f'{column}_Std']

indicadores_media_desvio.head()
```
